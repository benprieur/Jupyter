```python
# On s'intéresse aux liens éventuelles entre Nutriscore et Nova (application).
# Matrice de contingence.
```

```python
contigency = pd.crosstab(df_nutriscore_nova['nutriscore_grade'], df_nutriscore_nova['nova_group']) 
print(contigency)
```

```python
# On normalise la matrice.
```

```python
contigency_pct = pd.crosstab(df_nutriscore_nova['nutriscore_grade'], df_nutriscore_nova['nova_group'], normalize='index')
print(contigency_pct)
```

```python
plt.figure(figsize=(12,8)) 
sns.heatmap(contigency_pct, annot=False, cmap="YlGnBu")
```

```python
# On procède à un test du Khi2 (corrélation ou non ? <=> indépendance ou non).
```

```python
from scipy.stats import chi2_contingency
c, p, dof, expected = chi2_contingency(contigency_pct) 
print(p)
```

```python
# 0.9999694571775055 >> 0.01
# Le test n'est pas validé : pas d'indépendance ; nova et nutriscore sont corrélées.
```

```python
# Application : optimisation dans une catégorie alimentaire donnée
# Le Nutriscore uniquement.
# On affiche le premier résultat.
```

```python
codebarre = '0000000002257' #'0000000001281'
#print(codebarre)
#df_nutriscore_nova.head(50)

dfcode = df_nutriscore_nova[df_nutriscore.code==codebarre]
if (dfcode.shape[0] != 1):
    print("Pas de code barre correspondant")
    
product = dfcode.iloc[0]
#print(product.categories)
category = product.categories.split(', ')[0]
print(category)
nutriscore = product.nutriscore_grade
print(nutriscore)

dfresult = df_nutriscore_nova[df_nutriscore_nova['categories'].str.startswith(category, na=False)]
dfresult = dfresult[dfresult['nutriscore_grade'] < nutriscore]
dfresult.sort_values(by=['nutriscore_grade'], ascending=True)
dfresult.shape[0]
print ("--------------")
substitute = dfresult.iloc[0]
print (substitute)
```

```python
# On ajoute la colonne NutriNova.
```

```python
conditions = [
    (df_nutriscore_nova['nutriscore_grade'] == 'a'),
    (df_nutriscore_nova['nutriscore_grade'] == 'b'),
    (df_nutriscore_nova['nutriscore_grade'] == 'c'),
    (df_nutriscore_nova['nutriscore_grade'] == 'd'),
    (df_nutriscore_nova['nutriscore_grade'] == 'e')
    ]

values = [1, 2, 3, 4, 5]

import numpy as np
df_nutriscore_nova['NutriNova'] = np.select(conditions, values)
df_nutriscore_nova['NutriNova'] = df_nutriscore_nova['NutriNova'] + df_nutriscore_nova['nova_group']
df_nutriscore_nova.head()
```

```python
codebarre = '0000000002257' #'0000000001281'

dfcode = df_nutriscore_nova[df_nutriscore.code==codebarre]
if (dfcode.shape[0] != 1):
    print("Pas de code barre correspondant")
    
product = dfcode.iloc[0]
#print(product.categories)
category = product.categories.split(', ')[0]
print(category)
nutrinova = product.NutriNova
print(nutrinova)

dfresult = df_nutriscore_nova[df_nutriscore_nova['NutriNova'] < nutrinova]
dfresult.sort_values(by=['NutriNova'], ascending=True)
dfresult.shape[0]
print ("--------------")
substitute = dfresult.iloc[0]
print (substitute)
```
