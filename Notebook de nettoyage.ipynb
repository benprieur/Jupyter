```python
import pandas as pd
df = pd.read_csv("en.openfoodfacts.org.products.csv", sep='\t', error_bad_lines=False, usecols = ['code', 'categories', 'nutriscore_grade', 'nova_group', 'energy-kj_100g', 'carbohydrates_100g', 'sugars_100g', 'fat_100g', 'saturated-fat_100g', 'proteins_100g', 'fiber_100g','sodium_100g','alcohol_100g'], dtype='unicode')                                         
df.head(50)
```

```python
# On se base sur le nutriscore. 
#Quel est le % de lignes ayant un nutriscore ?
```

```python
a = df['nutriscore_grade'].isna().sum()
b = df.shape[0] 
c = ((b-a)/b)*100
print(c)
```

```python
# Ce sera notre dataset de référence (nutriscore défini).
# On créé ce dataset.
```

```python
df_nutriscore = df[ df['nutriscore_grade'].notna() ]
```

```python
# Combien de lignes dans notre dataset ?
```

```python
df_nutriscore.shape[0]
```

```python
# Des valeurs aberrantes dans notre colonne 'nutriscore' ?
# On vérifie. Si on trouve 717541 c'est que l'on a pas de valeurs aberrantes du nutriscore.
```

```python
dft = df_nutriscore.query("nutriscore_grade == 'a' or nutriscore_grade == 'b' or nutriscore_grade == 'c' or nutriscore_grade == 'd' or nutriscore_grade == 'e'")
dft.shape[0]
```

```python
# On va travailler sur les données nutritionnelles.
# Problème : quand il n'y a pas de valeur, elle est en NaN au lieu de 0.
# Le problème se pose pour les protéines, sel, graisses, glucides etc.
# Par exemple on cherche le nombre de lignes NaN des proteines.
```

```python
dftest = dft[ dft['proteins_100g'].isna() ]
dftest.shape[0]
```

```python
# On va remplacer par 0 tous les NaN de nos données nutritionnelles.
```

```python
dft['proteins_100g'] = dft['proteins_100g'].fillna(0)
dft['fat_100g'] = dft['fat_100g'].fillna(0)
dft['saturated-fat_100g'] = dft['saturated-fat_100g'].fillna(0)
dft['carbohydrates_100g'] = dft['carbohydrates_100g'].fillna(0)
dft['sugars_100g'] = dft['sugars_100g'].fillna(0)
dft['fiber_100g'] = dft['fiber_100g'].fillna(0)
dft['sodium_100g'] = dft['sodium_100g'].fillna(0)
dft['alcohol_100g'] = dft['alcohol_100g'].fillna(0)
dft.head(50)
```

```python
# On a des des valeurs singulières mais cohérentes : par exemple le sel en boîte est à 100 g de sodium.
# On commence par préciser des colonnes en float.
```

```python
dft = dft.astype({"proteins_100g":'float'})
dft = dft.astype({"fat_100g":'float'})
dft = dft.astype({"saturated-fat_100g":'float'})
dft = dft.astype({"carbohydrates_100g":'float'})
dft = dft.astype({"sugars_100g":'float'})
dft = dft.astype({"fiber_100g":'float'})
dft = dft.astype({"sodium_100g":'float'})
dft = dft.astype({"alcohol_100g":'float'})   
dft.shape[0]
```

```python
# On cherche les valeurs aberrantes, inférieures à zéro ou supérieures à 100.
```

```python
dftest = dft[dft['proteins_100g'].gt(100)]
#dftest.head(50)
dftest.shape[0]
```

```python
dft = dft[dft['proteins_100g'].le(100)]
dft = dft[dft['fat_100g'].le(100)]
dft = dft[dft['saturated-fat_100g'].le(100)]
dft = dft[dft['carbohydrates_100g'].le(100)]
dft = dft[dft['sugars_100g'].le(100)]
dft = dft[dft['fiber_100g'].le(100)]
dft = dft[dft['sodium_100g'].le(100)]

dft = dft[dft['proteins_100g'].ge(0)]
dft = dft[dft['fat_100g'].ge(0)]
dft = dft[dft['saturated-fat_100g'].ge(0)]
dft = dft[dft['carbohydrates_100g'].ge(0)]
dft = dft[dft['sugars_100g'].ge(0)]
dft = dft[dft['fiber_100g'].ge(0)]
dft = dft[dft['sodium_100g'].ge(0)]

dft.shape[0]
```
