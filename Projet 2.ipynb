```python
# https://medium.com/swlh/how-to-run-chi-square-test-in-python-4e9f5d10249d
contigency= pd.crosstab(df_nutriscore_nova['nutriscore_grade'], df_nutriscore_nova['nova_group']) 
print(contigency)
```

```python
contigency_pct = pd.crosstab(df_nutriscore_nova['nutriscore_grade'], df_nutriscore_nova['nova_group'], normalize='index')
print(contigency_pct)
```

```python
plt.figure(figsize=(12,8)) 
sns.heatmap(contigency_pct, annot=False, cmap="YlGnBu")
```

```python
from scipy.stats import chi2_contingency
c, p, dof, expected = chi2_contingency(contigency_pct) 
print(p)
```

```python
# 0.9999692943362465 >> 0.01
# Le test n'est pas validé : pas d'indépendance ; nova et nutriscore sont corrélées.
```

```python
df_nutriscore_sodium = df_nutriscore[ df_nutriscore['sodium_100g'].notna() ]
#df_nutriscore_sodium.head(50)
modalites = sorted(df_nutriscore_sodium['nutriscore_grade'].unique())
print(modalites)

groupes = []
for m in modalites:
    groupes.append(df_nutriscore_sodium[df_nutriscore_sodium['nutriscore_grade']==m]['sodium_100g'])
    
medianprops = {'color':"black"}
meanprops = {'marker':'o', 'markeredgecolor':'black',
            'markerfacecolor':'firebrick'}
    
plt.boxplot(groupes, labels=modalites, showfliers=False, medianprops=medianprops, 
            vert=False, patch_artist=True, showmeans=True, meanprops=meanprops)
plt.show()
```
